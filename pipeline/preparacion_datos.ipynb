### 1) Programas activos

El presente conjunto de datos informa el estado actual de los planes activos y los datos categoricos que corresponden a cada plan

```python
activos = pd.read_excel("../data/programas-activos.xlsx")
activos.head()
```

```python
activos.iloc[1]
```

La fecha de solicitud y fecha de rescindido crea atributos redudantes donde extraen el día, mes y año, no tomaremos en cuenta estos atributos inicialmente.

```python
new_activos = activos.drop(
    ["FECHA SOLICITUD Año", "FECHA SOLICITUD  Mes", "FECHA SOLICITUD  Día", "FECHA RESCINDIDO Año", "FECHA_RESCINDIDO - Mes", "FECHA_RESCINDIDO - Día"], axis=1)
```

La fecha de rescindido y la fecha de venta son un solo atributo que representan la fecha del estado actual donde solo en los casos en que un cliente tenga fecha de rescindido se reemplaza dicha fecha como fecha del estado.

```python
new_activos["fecha_estado"] = new_activos["FECHA DE VENTA"]
new_activos.iloc[3]["FECHA DE RESCINDIDO"]
```

```python
new_activos["fecha_estado"] = new_activos["fecha_estado"].where(
    new_activos["FECHA DE RESCINDIDO"] == "  ", new_activos["FECHA DE RESCINDIDO"])
new_activos.drop(["FECHA DE VENTA", "FECHA DE RESCINDIDO"], axis=1, inplace=True)
```

```python
new_activos.head(2)
```

No se evidencia registros vacios para la mayoría de atributos, sin embargo, si existen algunos para LOCALIDAD PROGRAMA, VALOR CUOTA y BARRIO PROGRAMA

```python
new_activos.describe(include="all")
```

Para localidad programa en caso de nulos usaremos la sucursal de administración

```python
new_activos["LOCALIDAD PROGRAMA"] = new_activos["LOCALIDAD PROGRAMA"].where(
    ~new_activos["LOCALIDAD PROGRAMA"].isna(), new_activos["SUCURSAL ADM"])
```

Para el valor de la cuota tomaremos el promedio del plan y lo aplicaremos al valor nulo

```python
means = new_activos[["VALOR CUOTA", "PLAN EXEQUIAL"]].groupby("PLAN EXEQUIAL").mean()
means
```

```python
new_activos[new_activos["VALOR CUOTA"].isna()] = new_activos[new_activos["VALOR CUOTA"].isna()].drop("VALOR CUOTA", axis=1).merge(means, on="PLAN EXEQUIAL")
```

Para BARRIO PROGRAMA, se tiene un valor llamado "Barrio No Identificado" el cual puede reemplazar los valores nulos

```python
new_activos["BARRIO PROGRAMA"] = new_activos["BARRIO PROGRAMA"].fillna(
    "Barrio No Identificado")
```

```python
new_activos = new_activos.rename(columns={"BARRIO PROGRAMA": "barrio_programa", "VALOR CUOTA": "cuota_programa", "LOCALIDAD PROGRAMA": "localidad_programa",
                                          "PLAN EXEQUIAL": "nombre_plan", "TIPO PLAN": "tipo_plan", "GENERO TOMADOR": "genero_cliente", "EDAD TOMADOR": "genero_cliente"})
new_activos.head(2)
```

Un mismo cliente puede tener varios programas, como es el caso de 5877 registros que se evidencian duplicados. Por tal razón, la llave primaría debería ser el id del programa.

```python
len(new_activos[new_activos["DOCUMENTO TOMADOR"].duplicated()])
```

```python
new_activos = new_activos.rename(
    columns={"PROGRAMA": "id_programa", "DOCUMENTO TOMADOR": "id_cliente"})
```

Eliminamos programas duplicados, estos solo estarán duplicados cuando incluyamos el historico

```python
new_activos.drop_duplicates(keep="first", subset="id_programa", inplace=True)
```

No se evidencia muchas diferencias entre sucursal venta y sucursal adm, adicionalmente hay más valores unicos en sucursal administrativa, por tal razón, se elimina sucursal venta. Hay registros que no tienen programa asociado se eliminan

```python
len(new_activos["SUCURSAL VENTA"].unique())
```

```python
len(new_activos["SUCURSAL ADM"].unique())
```

```python
new_activos[new_activos["SUCURSAL ADM"] !=
            new_activos["SUCURSAL VENTA"]].head(2)
```

```python
new_activos.drop(new_activos[new_activos["ESTADO"].isna()].index, inplace=True)
new_activos.drop("SUCURSAL VENTA", inplace=True, axis=1)
```

Finalmente renombramos los otros atributos para evitar espacios y/o diferencias entre mayusculas y minusculas

```python
new_activos = new_activos.rename(columns={"SUCURSAL ADM": "sucursal_adm", "ESTADO": "nombre_estado", "MOTIVO": "nombre_motivo", "TIPO MOTIVO": "tipo_motivo",
                            "DÍA DE PAGO MENSUAL": "día_pago_programa", "CANTIDAD INSCRITOS": "q_personas_programa", "CANTIDAD MASCOTAS": "q_mascotas_programa"})
new_activos.head()
```

### 2) Datos historicos 

El presente conjunto de datos brinda información de los planes que alguna vez fueron activados, su estado actual, así como su periodicidad de pago

```python
historico = pd.read_excel(
    "../data/Reporte-proyecto-prediccion-desistimiento-V3 2.xlsx")
historico.head(3)
```

```python
historico["Estados"].isna().sum()
```

```python
historico["Cuotas"].isna().sum()
```

```python
historico["Observaciones"].isna().sum()
```

```python
historico["GestionesRecaudo"].isna().sum()
```

Se evidencia que el identificador unico es codigo programa. Renombramos los atributos para que sea consecuentes con el conjunto de datos activos. Estados, Cuotas, Observaciones y Gestiones de recaudo se añadiran posteriormente si se requieren. Adicionalmente notamos un atributo más que es dirección 

```python
new_historico = historico.drop(
    ["Cuotas", "Observaciones", "GestionesRecaudo", "Unnamed: 9", "Estados"],axis=1).rename(columns={"CodigoPrograma":"id_programa", "Direccion": "direccion_cliente",
                                                                                   "LocalidadVenta": "localidad_programa", "EstadoActual": "nombre_estado",
                                                                                   "FechaRescindido": "fecha_estado"})
new_historico.drop([3965, 8921, 19881, 21855, 12646,
                   12839, 23979], inplace=True)
new_historico["fecha_estado"] = pd.to_datetime(new_historico["fecha_estado"])
```

Se observa que las columnas Estados, Cuotas, Observaciones y GestionesRecaudo son una lista de diccionarios, los cuales podrían convertirse en atributos individuales y si son varios elementos dentro de la lista en registros individuales.Lo dividiremos iniciando por los estados

```python
temp = historico.copy()[["CodigoPrograma", "Estados"]].dropna()
temp["Estados"] = temp["Estados"].str.replace(
    "[", "", regex=True).str.replace("]", "", regex=True)
temp = temp.set_index("CodigoPrograma")
temp = temp["Estados"].str.split("},", expand=True).stack(
).reset_index().drop("level_1", axis=1).set_index("CodigoPrograma")
temp = temp[0].str.split(",", expand=True)
temp[1] = temp[2].where(~temp[2].isna(), temp[1])
temp = temp.reset_index().rename(columns={"CodigoPrograma": "id_programa",0: "nombre_estado",
                1: "fecha_estado"}).drop(2, axis=1)
temp["nombre_estado"] = temp["nombre_estado"].str.replace(
    '{"Estado":"', "", regex=True).str.replace('"', "", regex=True)
temp["fecha_estado"] = temp["fecha_estado"].str.replace(
    '"fechainicio":"', "", regex=True).str.replace(
    '"fechacancelacion":"', "", regex=True).str.replace('"}', "", regex=True).str.replace('"', "", regex=True)
temp = temp.merge(new_historico.drop(
    ["nombre_estado", "fecha_estado"], axis=1), on="id_programa", how="left")
temp["fecha_estado"] = pd.to_datetime(temp["fecha_estado"])
temp
```

Concatenamos

```python
new_historico = pd.concat([new_historico, temp], axis=0, ignore_index=True).sort_values(
    "fecha_estado").drop_duplicates(subset=["id_programa", "nombre_estado"], keep="first")
```

```python
new_historico = new_historico.reset_index().drop("index", axis=1)
new_historico
```

Nombre estado tiene incluido también el nombre motivo

```python
temp = new_historico[["nombre_estado", "id_programa"]].set_index("id_programa")
temp = temp["nombre_estado"].str.split("-", expand=True).drop(2, axis=1)
temp = temp.reset_index().rename(
    columns={0: "nombre_estado", 1: "nombre_motivo", "index": "id_programa"})
temp
```

```python
new_historico = new_historico.drop("nombre_estado", axis=1).merge(temp, on="id_programa", how="left").sort_values(
    "fecha_estado").drop_duplicates(subset=["id_programa", "nombre_estado", "nombre_motivo"], keep="first")
new_historico
```

Nuestra base aumenta más de 20 mil registros, sin embargo, evidenciamos algunos valores vacios para la fecha_estado, muchos de ellos porque estan en activo activo, en cuyo caso sería la fecha de la factura, la cual no tenemos pero podríamos obtener de la base de datos de activos. Por otro lado, mantenemos el primer dato que no este duplicado asumiendo que fue la primera vez que solicito un retiro el usuario

```python
new_historico.describe(include="all")
```

Mas del 50% de esta base tiene nulos en la fecha_estado dado a que la moda es Activo. Se observa que la dirección de varios usuarios es Calle 50 # 24-34 que es la dirección de la empresa.

```python
new_historico.to_csv("../data/historico.csv", index=False)
```
